```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates

# Sample data
end_before= 2.90
end_before2= 4.7
y = np.array([40, 40, 40])
y2 = np.array([30,30,30])
y3 = np.array([20,20,20])
y4 = np.array([10,10,10])
x = np.linspace(1,2.7, num=len(y))
x2 = np.linspace(1.5, end_before, num=len(y))
x3 = np.linspace(3.1, end_before2, num=len(y))
x4 = np.linspace(4.1, 7, num=len(y))


# Create the plot
fig, ax = plt.subplots(figsize=(15, 6))
plt.subplots_adjust(bottom=0.30, top=0.85)
line1, =ax.plot(x, y, '-', color='#B08AD5', linewidth=25, solid_capstyle = 'round')
line2,= ax.plot(x2, y2, '-', color='#6FA8DC', linewidth=25, solid_capstyle = 'round')
line3, = ax.plot(x3, y3, '-', color='#7FB3A2', linewidth=25, solid_capstyle = 'round')
line4, = ax.plot(x4, y4, '-', color='#E6A57A', linewidth=25, solid_capstyle = 'round')

#Color Background
fig.patch.set_facecolor('black')
ax.set_facecolor('black')
ax.tick_params(colors='White')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.title.set_color('white')
for spine in ax.spines.values():
    spine.set_color('white')


# Add text directly on the curves
ax.text(x[1], y[1], 'YT Music Library + Data Foundation', color='black', fontsize=10, ha='center', va='center')
ax.text(x2[1], y2[1], 'Listening History Data Product', color='black', fontsize=10, ha='center', va='center')
ax.text(x3[1], y3[1], 'ETL Automation & Observability', color='black', fontsize=10, ha='center', va='center')
ax.text(x4[1], y4[1], 'Non-Proprietary Database Migration & API Connectivity — Early Concept', color='black', fontsize=10, ha='center', va='center')

# Set the y-axis limits (optional, but good practice)
ax.set_ylim(-10, 50)

# Define numerical tick locations on the y-axis
y_tick_locations = [0, 10, 20, 30, 40, 50]
x_tick_locations = np.arange(7)

# Define the custom labels corresponding to these locations
y_tick_labels = ['', 'Product D', 'Product C', 'Product B', 'Product A', '']

y_tick_colors = [
'white',        # 0 (vide)
'#E6A57A',      # Product D
'#7FB3A2',      # Product C
'#6FA8DC',      # Product B
'#B08AD5',      # Product A
'white'         # 50 (vide)
]

x_tick_labels = ['11/25', '12/25', '01/26', '02/26' , '03/26', '04/26', '05/26']


# Apply the custom tick locations and labels
ax.set_yticks(y_tick_locations)
ax.set_yticklabels(y_tick_labels)

# Apply custom colors to y-tick labels
for label, color in zip(ax.get_yticklabels(), y_tick_colors):
    label.set_color(color)
    label.set_fontweight('bold')

ax.set_xticks(x_tick_locations)
ax.set_xticklabels(x_tick_labels)

#Create a new line for phase with vertical markers
phases = [
    ("Phase 1 – Foundation", 1.4),
    ("Phase 2 – Insights", 2.5),
    ("Phase 3 – Scale", 4.5),
]

phase_lines = [
    ("Phase 1 – Foundation", 0.8, 2.0),
    ("Phase 2 – Insights", 2.0, 3),
    ("Phase 3 – Scale", 3, 6.5),
]
for label, x_start, x_end in phase_lines:
    # Secondary line
    ax.hlines(y=-1.5, xmin=x_start, xmax=x_end,
              color='white', linewidth=2, alpha=1)

    # Vertical markers
    ax.vlines(x_start, 3, -1.6, color='white', linewidth=2, alpha=1)
    ax.vlines(x_end,   3, -1.6, color='white', linewidth=2, alpha=1)
for text, xpos in phases:
    ax.text(
        xpos, -3,
        text,
        color='white',
        fontsize=10,
        ha='center',
        va='top',
        alpha=0.85,
        fontweight='medium'
    )
# Add a general y-axis label if needed
ax.set_ylabel('Sprints',fontweight='bold')

plt.title('Roadmap Youtube Music analytics',fontweight='bold')
plt.xlabel('Dates',fontweight= 'bold',)
plt.grid(False)

# Fig save
plt.savefig(
    "Roadmap YT music Analytics.png",
    dpi=300
)

# Display the plot
plt.show()
```
